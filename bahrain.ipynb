# 2023 Bahrain GP Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import fastf1 as ff1
```

```python
ff1.Cache.enable_cache('cache')
session = ff1.get_session(2023, 'Bahrain', 'FP1')
session.load()
```

```python
#checking that data loaded correctly by printing the list of drivers
drivers = pd.unique(session.laps['Driver'])
print(drivers)
```

```python
#creating a dataframe with the drivers and their fastest lap times for further analysis
from fastf1.core import Laps
fastest_laps_list = []
for driver in drivers:
    fastest_lap = session.laps.pick_driver(driver).pick_fastest()
    fastest_laps_list.append(fastest_lap)

fastest_laps = Laps(fastest_laps_list).sort_values(by='LapTime').reset_index(drop=True)
fastest_laps[['Driver', 'LapTime']]
```
